### 模块与包简介

#### 模块
- 当代码较多时，将其放在一个文件里时难以维护，所以将其拆分为多个.py文件，以便使用与维护，同时解决命名空间（重复）的问题
- 当各自的模块下面有相同的变量命名时，由于引用时所用的模块的名称各不相同，所以不会混乱


#### 包
- 当模块较多时，也容易重名，可以将不同的模块放入不同的包中，解决模块冲突的问题。
- 在文件系统中，包就是文件夹，而模块就是.py文件，包可以有多级


### 导入模块
- 你可以认为模块名就是一个指向已导入模块的变量，通过该变量，我们可以访问math模块中所定义的所有公开的函数、变量和类
- 导入模块的格式：
    - 格式1：import module1[,module2,....modulen] 
    - 格式2： **注意：此格式中程序中的变量会将模块中的变量覆盖**
        - 导入有限命名空间：from modeul import name1[,name2...namen]  
        - 导入全部命名空间：from modeul import *  

- 引入自定义模块   import 模块名（不用加.py）
- 一个模块只能引入一次，及时执行多次引用
- 第三方模块

```python
#你可以认为math就是一个指向已导入模块的变量，通过该变量，我们可以访问math模块中所定义的所有公开的函数、变量和类：
import math
print(math.pow(2, 0.5)) # pow是函数
print(math.pi) # pi是变量
```

```python
# 导入某几个函数（变量）
from math import pow, sin, log
print(pow(2, 10),sin(3.14))
```

- 如果遇到名字冲突
    - 导入时：import module1[,module2,....modulen] ， 则使用时通过模块名引用函数名 (模块名.函数)
    - 直接在导入是改名称：from  modeul import 变量名 as 新变量名称   # 

```python
#如果使用import导入模块名，由于必须通过模块名引用函数名
import math, logging
print (math.log(10),logging.log(10, 'something'))
```

```python
from math import log as cc
cc(3)
```
