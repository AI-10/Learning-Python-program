## 封装 继承 多态

### 继承

#### 基本语法规则

```python
#父类
class Person(object):
    def __init__(self, name, gender):
        self.name = name
        self.gender = gender
        
#定义Student类时，只需要把额外的属性加上，例如score：
class Student(Person):
    def __init__(self, name, gender, score):
        Person.__init__(self.name,gender) #也可以用super(Student, self).__init__(name, gender)，super()是获取父类的定义，而不是父类的对象
        self.score = score
#一定要在构造函数中调用父类的构造函数，否则，继承自 Person 的 Student 将没有 name 和 gender。
```

#### 实例全面解析继承

```python
#类定义
class People:
    
    #定义基本属性
    name = ''
    age = 0
    
    #定义私有类属性,私有属性在类外部无法直接进行访问,可以间接访问
    __high = 0
    
    #定义构造方法
    def __init__(self,n,a,w):
        self.name = n
        self.age = a
        self.__weight = w#定义实例私有属性
        
    def speak(self):
        print("%s 说: 我 %d 岁。" %(self.name,self.age))
        
    def say(self):
        print("*****")
        
# 单继承示例
class student(People):# 子类继承父类的除构造方法之外的所有成员（包括（私有）属性、方法）
    grade = ''
    
    def __init__(self,n,a,w,g):
        #调用父类的构函
        People.__init__(self,n,a,w)
        self.grade = g
        
    #重写父类的方法
    def speak(self):
        print("%s 说: 我 %d 岁了，我在读 %d 年级"%(self.name,self.age,self.grade))
        
# 多继承示例
# 多继承有父类相同的方法时，将从左至有检索方法。
 
print(Student.mro()) #显示继承层次关系

 
s = student('ken',10,60,3)
s.say()  # 调用父类的方法
s.speak()# 调用子类重写的方法

print(dir(s),dir(object))# 显示属性（可见s继承了父类的所有属性，私有属性__high,self.__weight也被继承，只是不能被直接调用）
# 调用父类的私有属性（用dir显示的属性格式调用）
print(s._People__high)
print(s._People__weight)
```
