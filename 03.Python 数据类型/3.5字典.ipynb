# 字典 dict
- 无序组合数据，键值对(key - value)映射的形式存贮

```python
help(dict)
```

### 字典的特性
- 字典是序列类型，但字典内部的数据是无序的，所以无法索引和分片
- 所有的元素都是kv对，不同于列表元组等数据类型，字典的索引是用户自定义的
- key必须是不可变对象（如str，int，float，tuple等，而不能是字典，集合，列表），而且一个字典的key要唯一，value可以是任何值

### 创建字典
- 空字典，有值字典

```python
"""
创建空字典
"""
#直接创建
d = {}
#dict函数创建
s = dict()
print(d,s,type(d),type(s))
```

```python
"""
创建有值字典
"""
d = {"one":1,"two":2,"three":3,"fore":4}
print(d,type(d),id(d))

d = dict({"one":1,"two":2,"three":3,"fore":4})
print(d,type(d),id(d))

#dict函数+关键字参数创建
d = dict(one=1,two=2,three=3,fore=4)
print(d,type(d),id(d))

#dict函数+数组创建
d = dict({("one",1),("two",2),("three",3),("fore",4)})
print(d,type(d),id(d))
d = dict([("one",1),("two",2),("three",3),("fore",4)])
print(d,type(d),id(d))
```

### 字典的相关操作

```python
"""
字典元素的访问
"""
# 字典可以通过键进行索引
#- 格式1： 字典名 [key]
#- 格式2： 字典名.get [key]
d = {"one":1,"two":2,"three":3,"fore":4}
s = d["one"]      # 根据key取值，无此key时会报错
f = d.get("hhh")  # 当没有此key时返回None，而不报错
print(d.get("one"),100)  #找到key值，并返回默认值100
print(s,d)
```

```python
"""
元素的添加、赋值与修改
"""
#- 格式： 字典名 [ key ] = 赋值数据
#- 当没有此key时，添加此键值对，当有次key时，就会修改原key对应的值
d = {"one":1,"two":2,"three":3,"fore":4}
d["five"] = 5
d ["one"] = 12
print(d)
```

- for循环
        1. for循环+键值遍历
        2. for循环+keys函数遍历
        3. for循环+values函数遍历
        4. for循环+items函数遍历

```python
"""
字典的遍历（迭代）
"""
#- for循环
#        1. for循环+键值遍历
#        2. for循环+keys函数遍历
#        3. for循环+values函数遍历
#        4. for循环+items函数遍历#
#字典的遍历
d = {"one":1,"two":2,"three":3,"fore":4}
for k in d:             #遍历k
    print(k,d[k])
for k in d.keys():      #遍历k
    print(k,d[k])
for k in d.values():    #遍历v
    print(k)            # values() 方法实际上把一个 dict 转换成了包含 value 的list。
for k,v in d.items():   #遍历k和v
    print(k,v)          #items() 方法把dict对象转换成了包含tuple的list，我们对这个list进行迭代，可以同时获得key和value：
for k,v in enumerate(d):#枚举k
    print(k,v)    
```

### 字典相关函数与方法

```python
"""
字典的删除
"""
# del 字典名 [key]
# 字典名.pop(key)
d = {"one":1,"two":2,"three":3,"fore":4}
s = len(d)
del d ["one"]
d.pop("fore")
print(d,s)
```
