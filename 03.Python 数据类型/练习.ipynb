```python
def  getnum():
    num=[]
    inum = input("请输入数字")
    while inum != "":
        num.append(eval(inum))
        inum = input("请输入数字")
    return num
```

```python
def means(num):
    s = 0
    for i in num:
        s += i
    return s/len(num)
```

```python
n = getnum()
```

```python
print(n)
```

```python
def get():
    txt = open(r"C:\Users\Administrator\Desktop\test.txt","r",encoding ="UTF-8").read()
    txt = txt.lower()
    for i in '!"#$%&()*+,-./:;<=>?@[\\]^_‘{|}~':
        txt = txt.replace(i," ")
    return txt
```

```python
h = get()
wh = h.split()
counts = {}
for i in wh:
    counts[i] = counts.get(i,0)+1
items = list(counts.items())
items.sort(key = lambda x:x[1],reverse = True)
items
```
