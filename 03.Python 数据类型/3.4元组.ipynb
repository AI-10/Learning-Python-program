# 元组  tuple
- 可以看做是不可以更改(内容)的list  (一旦创建，就无法更改其元素）
- help(tuple) 

help(tuple) 

### 元组的特性
- 是序列表，有序
- 可以访问，但不可以修改，数据比较安全
- 元组数据可以是任何类型

### 创建元组
 - 创建空元组,单值元组，多值元组，其他结构创建元组。

```python
"""
空元组时：t = ()
"""
#创建空元组
t = ()
print(t,type(t))
```

```python
"""
单个值时：t = (x,)   
注意：逗号必须要加，消除定义一个数的歧义。
"""
#单值元组的括号与逗号情况
t = (1)#没有逗号是int类型
x = 1,#没有括号是元组类型
y = (1,)#加入逗号变成元组类型
z = 1 #赋值,int类型
print(t,type(t))
print(x,type(x))
print(y,type(y))
print(z,type(z))
```

```python
"""
多个值时：t = (x1,x2,x3,....)
元组中的数据类型可以不同
"""
#有无括号和最后有无逗号不影响数据类型，单推荐使用格式：y = (1,2,3,4)
t = (1,2,3,"w",)
y = (1,2,3,4)
z = 1,2,3,4
x = 1,2,3,4,
print(t,type(t))
print(y,type(y))
print(z,type(z))
print(x,type(x))
```

### 元组的相关操作
- list的特性，除了原地修改，其他的元组都具有，比如索引，分片，序列相加，成员操作等
- count()  len()  max()  min()

#### （1）元组的访问

```python
# 取值 ：元组名[下标]
l = (1,2,3,4,5,2,3)
i = l[3] # 不超范围
print(i)
i = l[-5]
print(i)
# 元组的不可修改性
l = (1,2,3,4,[5,2,3])
#l[-1]= 100
l[-1][2]= 100  # 注意不可改变的是元组的元素，此处修改的是元组里的list的元素，元组的list元素可变，因为list里面的元素存贮在元组内存外
print(l)
```

#### (2)删除元组

```python
l = (1,2,3,4,[5,2,3])
del l
print(l)
```

#### (3)元组相加 数乘

```python
a = (1,2,3,4)
b = (5,2,3)
c = a + b
print(c)
d = c * 3
print(d)
print(id(a),id(b),id(c),id(d))
```

#### （4）类型转换

```python
a = [1,3,4,5,3,5]
b = tuple(a)
print(b)
```

#### （6）元组变量交换法

```python
#元组变量交换法
a = 1
b = 2
#java交换法
c = a
a = b
b = c
print(a,b)
#py交换法
a,b = b,a
print(a,b)
```
