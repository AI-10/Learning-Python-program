# 变量 
  - 概述：
    - 变量就是程序可操作的存储空间的名称(在存储数据时，先在内存开辟一块空间，这块内存空间就是定义的变量空间，变量=10，就是在变量空间里存入10)
    - 或者程序运行期间可改变的数据
    - 每个变量有特定的类型 
  - 作用：
    - 将不同类型的数据存入内存

- 当我们写：a = 'ABC'时，Python解释器干了两件事情：
    - 1. 在内存中创建了一个'ABC'的字符串；
    - 2. 在内存中创建了一个名为a的变量，并把它指向'ABC'

```python
a = 'ABC'
b = a
a = 'XYZ'
print (b)
```

- 可以把一个变量a赋值给另一个变量b，这个操作实际上是把变量b指向变量a所指向的数据，例如上面的代码：


```python
"""
变量的定义
"""
#定义一个变量  变量名 = 初识值（初始值可以确定变量的类型）
age = 0

"""
数据的存储
"""
#存储数据      变量名 = 数据值
age = 10#覆盖先前的数据
print(age)#使用变量时要首先定义
"""
变量的删除
"""
#删除变量  del 变量名
#del age
print(age)
"""
变量的类型
"""
#查看变量的类型 type(变量名)
age = 10
print(type(age))
"""
变量的地址
"""
#查看变量的地址  id(变量名)
print(id(age))
```

# 赋值

**1. 什么是赋值？其基本形式是什么？**
 -  赋值就是用赋值语句把对象（参考对象一章）赋给一个名称，其目的是建立对象的引用值
 - 基本形式  赋值目标（变量名、对象元素） = 要赋值的对象

**2. 赋值语句的特性是什么？**
  - 赋值语句建立对象引用值（py赋值语句把对象引用值存贮在变量名或数据结构的元素中，而不是复制对象）
  - 变量名首次赋值便被创建（一旦赋值，当变量名出现在表达式中时，就会被其引用的值所替代）
  - 变量引用前必须先赋值
  - 隐形赋值的操作（赋值语句会在许多情况下使用，如导入模块、函数和类的定义等）     

**3. 赋值语句的形式有哪些？**
  * 基本形式               A = "SB" 
  * 元组赋值（位置性）     A,B = "SB","SD"
  * 列表赋值（位置性）     [A,B] = ["SB","SD"]
  * 序列赋值（通用性）     A,B,C,D = "abcd" ==> A="a" B="b" C="c" D="d"
      - *元组和列表的统一---任何变量名的序列可以赋值给任何值的序列，而Py会按位置一次赋值一个元素，序列类型可混合，但长度要匹配*  
      - *语句执行时，Python会临时建立一个元组来储存右侧变量原始的值，分解语句也是交换两变量的值，却不需要自行建立临时变量的方式*
  * 扩展的序列解包 A,* B = "abcd"==>A = "a" B = "bcd"
      - *第一个字母匹配a 剩下的匹配bcd*
  * 多重目标拷贝 A = B = "sd"
  * 增强赋值语句（运行快）A += 42 ==> A = A + 42

```python
# 赋值语句形式
A = "SB"
print(A)

A,B = "SB","SD"
print(A,B)

[A,B] = ["SB","SD"]
print(A,B)

A,B,C,D = "abcd" 
print(A,B,C,D)

A,* B = "abcd"
print(A,B)

A = B = "sd"
print(A,B,id(A),id(B))#地址不变

A += "l"
print(A)
```
