### 一、一个温度转换实例

```python
# 温度转换代码实现
T = input ("请输入带有单位的温度值:")
if   T[-1] in  ["f" ,"F"]  :
    C = (eval(T[0:-1])-32)/1.8
    print ("转换后的温度为{:.2f}C".format(C))
elif T[-1] in  ["c" , "C"] :
    F = 1.8*eval(T[0:-1])+32
    print ("转换后的温度为{:.2f}F".format(F))
else:
    print("输入格式错误")
```

###  二、Python 编程 IPO思想
- **I - input 输入**:带有温度单位的温度值
- **P- process 处理**:根据温度标志进行转换
- **O- output输出**:转化后带有单位的温度值

### 三、代码元素解析

#### 1. 程序格式框架
- Python的格式框架是由缩进实现的，可多层缩进。
- 缩进的要求
    - 严格明确:缩进是语法的一部分，缩进不正确程序运行错误
    - 所属关系:缩进是Python表达代码间包含和层次关系的唯一手段
    - 长度一致:未规定缩进长度，但程序内一致即可，一般用4个空格或1个TAB

#### 2.注释
- 注释是不被执行的辅助性说明信息
- Python的注释符：
     - 单行： #
     - 多行："""  """

#### 3.变量、变量赋值与变量命名
- 变量：程序中用于保存和表示数据的占位符号，如上述代码中的T,C,F
- 变量赋值：可以使用等号(=)向变量赋值或修改值， =被称为赋值符号，如上述代码中的：C = (eval(T[0:-1])-32)/1.8就是给C赋值
- 变量命名规则：
    - 大小写字母、数字、下划线和中文等字符及组合
    - 大小写敏感、首字符不能是数字、不与保留字相同

#### 4.Python保留字
- 保留字是被编程语言内部定义并保留使用的标识符，这些字符串不能用来定义变量
- Python中的33个保留字


#### 5.数据类型
- 上述代码中出现的数据类型有：数字，字符串，列表
    - 数字：包括整数和浮点数
    - 字符串：由0个或多个字符组成的有序字符序列，用引号表述
    - 列表：由0个或多个数据组成的有序序列，用 [ ] 表述
- 当然还有其他数据类型，具体内容请看后面章节

#### 6.语句和函数：
